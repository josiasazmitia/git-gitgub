# Ejemplo de uso de Magic Cells en Jupyter Notebook

```python

# %matplotlib inline para gráficos dentro del notebook
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('números ejemplo')
plt.show()
    
```

```python

# %%time para medir el tiempo de ejecución de la celda
%%time
suma = 0
for i in range(100000):
    suma += i
print(suma)
    
```
